# Palindriome
when a string is read in a reverse order and we get the same value as we 
get when we read in start, then that string is called palindrome 
e.g MADAM , read this from left to right and right to left that will give us same result


```python
a = input("Enter any name here: ")
b= a[-1::-1] # start_val:stop_val:step_val, double colon means we are specifying step value 
if a==b:
    print('Its a palindrome string')
else:
    print('It is not a palindrome string')
```

# Machine learning 

```python
#importing wine data from sklearn 
from sklearn.datasets import load_wine
```

```python
wine_data = load_wine()
```

```python
dir(wine_data) #This will tell what all attributes does our loaded data has 
# which can be called using dot syntax 
```

```python
X= wine_data.data
y= wine_data.target
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
print(y_test.shape)
print(X_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
my_model =LogisticRegression()
```

```python
my_model.fit(X_train,y_train)
```

```python
prediction = my_model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test,prediction)
accuracy
```

# Naive Bayes

```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
os.listdir()
```

```python
data = pd.read_csv('diabetes.csv')
data.head(3)
```

```python
y= data.Outcome
X= data.drop(['Outcome'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.33,random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
NB_model = GaussianNB()
```

```python
NB_model.fit(X_train,y_train)
```

```python
pred_nb = NB_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
accuracy = accuracy_score(y_test,pred_nb)
accuracy
```

```python
conf_mat = confusion_matrix(y_test,pred_nb)
conf_mat
```
