# Logistic Regression 

```python
from sklearn.linear_model import LogisticRegression
```

```python
y= data.target
X= data.drop("target",axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.30,random_state=0)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
my_model = LogisticRegression()
```

```python
my_model.fit(X_train,y_train)
```

```python
pred_LR = my_model.predict(X_test)
```

```python
#checking accuracy 
from sklearn.metrics import accuracy_score,confusion_matrix
```

```python
acc_LR = accuracy_score(y_test,pred_LR)
acc_LR
```

```python
conf_mat_LR = confusion_matrix(y_test,pred_LR)
conf_mat_LR
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
my_DT = DecisionTreeClassifier(max_depth=7)
```

```python
my_DT.fit(X_train,y_train)
```

```python
pred_DT = my_DT.predict(X_test)
pred_DT
```

```python
acc_DT = accuracy_score(y_test,pred_DT)
acc_DT
```

```python
conf_mat = confusion_matrix(y_test,pred_DT)
conf_mat
```

```python
#plotting Decision Tree
from sklearn import tree
```

```python
plt.figure(figsize=(15,10))
tree.plot_tree(my_DT,filled=True)
```
